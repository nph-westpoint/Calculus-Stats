# Linear Interpolation of time series data
The code from git hub can be used to import the necessary objects and functions to impute the data. This sheet will help explain how.

```python
from tsl_imputer import (
    linear_interpolation,
    get_file,
    Imputer)
import pandas as pd
import numpy as np
```

### Imputing the data - step by step
In the following code, we take you through the process so you can see what the data looked like before and after.<br><br>
Select the csv file "test_all_groups.csv" to get the results talked about in the explanation below.

```python
admin_cols = 2
filename = get_file()
df = pd.read_csv(filename)
imp = Imputer(df,admin_cols)
imp.data
```

As you can see with the test file, there are several rows of data (10 to be exact) that are missing data. The constructor method for the Imputer actually creates two data structures with the rows and columns. Let's check them out:

```python
imp.rows
```

This means that there is data missing in rows 2,3,5,7,... and so on. You can verify that by looking at the printed out data frame.<br><br>
Now lets see the columns that are missing data for each of the listed rows...

```python
imp.cols
```

This means that row 2 has missing data in column 6, row 3 has data missing in columns 0 and 1, row 5 has data missing in columns 2 and 6, and so on.

We now are set to impute data or throw out a row if it does not meet the criteria. Nothing has been done yet.

```python
imp.imputed_data
```

```python
imp.impute_data()
```

Once the method `impute_data` is run, the output is the new data frame with the rows that passed the conditions of being imputed (less than 3 missing data and no negative values).

```python
# Note that this is not the method, this data frame is stored in the imp object and can be used later.
imp.imputed_data
```

```python
imp.deleted_rows
```

Note that rows 3,5, and 7 were deleted by the algorithm because one of the conditions were not met.

As an alternative, one can simply run the linear_interpolation function and it will do all of these steps for you.

```python
admin_cols = 2
new_data = linear_interpolation(admin_cols)
new_data
```
